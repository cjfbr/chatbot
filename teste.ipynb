```python
from src.question_parser import parse_question
from src.query_engine import *
from src.data_loader import load_data
from src.response_generator import generate_response

data = load_data()

questions = [
    # 🏛 Current wage (10)
    "What is the minimum wage in Texas?",
    "How much do workers earn per hour in California?",
    "Tell me the current minimum wage in New York.",
    "What’s the pay rate for Oregon employees?",
    "Current wage in Florida?",
    "How much is the minimum wage in Washington state?",
    "Wage in Alabama right now.",
    "What do people get paid per hour in Nevada?",
    "Show me the minimum wage for Illinois.",
    "How much can someone earn in Colorado this year?",

    # 📈 Maximum & Minimum (10)
    "Which state has the highest minimum wage?",
    "Where is the minimum wage the lowest?",
    "Which state pays workers the most?",
    "Which state has the smallest minimum wage?",
    "Show the state with the biggest minimum wage.",
    "Where is the top-paying state for minimum wage?",
    "Which state has the least wage per hour?",
    "What is the largest minimum wage among all states?",
    "Who pays the highest hourly minimum wage in the U.S.?",
    "Which state has the lowest legal minimum pay?",

    # 💵 Tipped workers (7)
    "What is the tipped minimum wage in Texas?",
    "How much do tipped employees earn in Florida?",
    "Which state has the highest tipped wage?",
    "What’s the smallest tipped wage in the country?",
    "Compare tipped wage between California and Texas.",
    "Tipped worker wage history in New York.",
    "Show the tipped wage rate in Nevada.",

    # 👶 Age / minors (7)
    "Minimum wage for minors in California.",
    "What do children earn in Florida?",
    "Pay rate for kids working in Texas.",
    "Wages for teenagers in New York.",
    "What’s the minimum age-based pay in Illinois?",
    "Wage for minors with certificate in Washington.",
    "How much can a child legally earn per hour?",

    # 📜 Historical wage (8)
    "What was the minimum wage in Texas in 2010?",
    "Wage in California in 2015.",
    "How has the wage changed in Washington over the years?",
    "When did New York raise its minimum wage?",
    "Minimum wage history for Florida.",
    "Show wage evolution for Oregon since 2000.",
    "How did the minimum wage change in Illinois in 2020?",
    "Yearly wage trend for Alaska.",

    # ⚖️ Comparison (8)
    "Compare wage in Texas and California.",
    "Wage difference between Washington and Oregon.",
    "Which state pays more, Nevada or Arizona?",
    "Compare minimum wage in Florida vs Georgia.",
    "Who has the higher wage, New York or New Jersey?",
    "Wage comparison between Alaska and Hawaii.",
    "Compare wages for tipped workers in Texas and California.",
    "Which state offers better pay, Colorado or Utah?"
]
 # paste the 50 here
for q in questions:
    parsed = parse_question(q)
    q_type = parsed["type"]
    result = None

    if q_type == "max":
        result = query_max(data)
    elif q_type == "min":
        result = query_min(data)
    elif q_type == "max_tipped":
        result = query_max_tipped(data)
    elif q_type == "min_tipped":
        result = query_min_tipped(data)
    elif q_type == "compare":
        result = query_compare(data, parsed["states"])
    elif q_type == "current":
        result = query_current(data, parsed["states"][0] if parsed["states"] else None)
    elif q_type == "history":
        result = query_history(data, parsed["states"][0] if parsed["states"] else None, parsed["year"])
    elif q_type == "tipped":
        result = query_tipped(data, parsed["states"][0] if parsed["states"] else None)
    elif q_type == "age":
        result = query_age(data, parsed["states"][0] if parsed["states"] else None)

    answer = generate_response(parsed, result)
    print(f"Q: {q}\nA: {answer}\n{'-'*50}")
```
